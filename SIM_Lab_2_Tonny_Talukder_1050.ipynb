```
import numpy as np

# Generate two 3Ã—3 matrices with random integers between 1 and 10
A = np.random.randint(1, 11, (3, 3))
B = np.random.randint(1, 11, (3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

```

```
# Matrix addition
add_result = A + B

# Matrix subtraction
sub_result = A - B

# Matrix multiplication (dot product)
mul_result = np.dot(A, B)

print("A + B:\n", add_result)
print("A - B:\n", sub_result)
print("A x B:\n", mul_result)
```

```
# Determinant
det_A = np.linalg.det(A)

# Rank
rank_A = np.linalg.matrix_rank(A)

# Inverse (only if determinant is not zero)
try:
    inv_A = np.linalg.inv(A)
except np.linalg.LinAlgError:
    inv_A = "Matrix A is singular and does not have an inverse."

print("Determinant of A:", det_A)
print("Rank of A:", rank_A)
print("Inverse of A:\n", inv_A)
```

```
import numpy as np
import matplotlib.pyplot as plt

# Create two random 4x4 matrices with integer values between 0 and 10
A = np.random.randint(0, 10, size=(4, 4))
B = np.random.randint(0, 10, size=(4, 4))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix operations
matrix_sum = A + B              # Addition
matrix_diff = A - B             # Subtraction
matrix_prod = np.dot(A, B)      # Multiplication (dot product)

# Flatten matrices to 1D arrays for bar plotting
sum_flat = matrix_sum.flatten()
diff_flat = matrix_diff.flatten()
prod_flat = matrix_prod.flatten()

# Create subplots for visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot Addition
axes[0].bar(range(1, 17), sum_flat, color='blue')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Element Index')
axes[0].set_ylabel('Value')

# Plot Subtraction
axes[1].bar(range(1, 17), diff_flat, color='green')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Element Index')
axes[1].set_ylabel('Value')

# Plot Multiplication
axes[2].bar(range(1, 17), prod_flat, color='red')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Element Index')
axes[2].set_ylabel('Value')

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a 4x4 matrix with random values between 0 and 1
matrix = np.random.rand(4, 4)  # Values in [0, 1)
print("Random 4x4 Matrix:\n", matrix)

# Create row and column labels
row_labels = ['Row 1', 'Row 2', 'Row 3', 'Row 4']
col_labels = ['Col 1', 'Col 2', 'Col 3', 'Col 4']

# Plot the heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(matrix, annot=True, cmap='YlGnBu', linewidths=0.5, linecolor='black',
            xticklabels=col_labels, yticklabels=row_labels)

plt.title('Heatmap of 4x4 Random Matrix')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Create two vectors with 15 random floats
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

# Create an index for the x-axis
x = np.arange(15)

# Plot the vectors
plt.plot(x, vector1, marker='o', linestyle='-', label='Vector 1')
plt.plot(x, vector2, marker='s', linestyle='--', label='Vector 2')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Two Random Vectors')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
