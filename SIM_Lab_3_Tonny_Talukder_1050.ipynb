```python
import numpy as np
import matplotlib.pyplot as plt

# 1×10 vector of uniform random numbers
uniform_random_vector = np.random.rand(10)
print("1x10 Uniform Random Vector:\n", uniform_random_vector)

# Histogram plot
plt.hist(uniform_random_vector, bins=5, edgecolor='black')
plt.title('Histogram of 1×10 Uniform Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 500 normal distributed values
normal_random_values = np.random.normal(0, 1, 500)

# Histogram plot
plt.hist(normal_random_values, bins=30, edgecolor='black')
plt.title('Histogram of Normally Distributed Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Set seed for reproducibility
np.random.seed(10)
seeded_matrix = np.random.randint(1, 51, (3, 3))
print("Seeded Random Matrix:\n", seeded_matrix)
```

```python
# Random matrix without seed
np.random.seed(None)
non_seeded_matrix = np.random.randint(1, 51, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)
```
